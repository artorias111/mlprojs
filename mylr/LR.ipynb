```python
from pydataset import data
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
pima=data('Pima.tr')
```

```python
pima.plot(kind='scatter',x='skin',y='bmi')
plt.show()
```

```python
#test train split (supervised learning)
x_train,x_test,y_train,y_test=train_test_split(pima.skin,pima.bmi)
```

```python
plt.scatter(x_test,y_test,label='training data',color='r',alpha=.7)
plt.scatter(x_train,y_train,label='testing data',color='b',alpha=.7)
```

```python
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1,1),y_train.values)
```

```python
pred=lr.predict(x_test.values.reshape(-1,1))
```

```python
plt.plot(x_test,pred,label='linear regression')
plt.scatter(x_test,y_test,label='test data',color='r')
```

```python
lr.score(x_test.values.reshape(-1,1),y_test.values)
```
